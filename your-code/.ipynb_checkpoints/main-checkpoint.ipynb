#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
    url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, "html5lib")
````

````python
a = soup.find_all('a', {'class': 'image'})
a
links = []
pattern = 'File:.+'
for element in a:
    a_name = element.get('href')
    link_fragment = re.findall(pattern, a_name)
    link = f'https://en.wikipedia.org/wiki/Walt_Disney#/media/{link_fragment[0]}'
    links.append(link)
links
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, "html5lib")
````

````python
a = soup.find_all('a')

links = []
for element in a:
    try:
        if(element['href'].startswith('http')):
            links.append(element['href'])
        elif (element['href'].startswith('#')):
            links.append('https://en.wikipedia.org/wiki/Python' + element['href'])
        else:
            links.append('https://en.wikipedia.org' + element['href'])
    except:
        pass
            
links
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code 
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
tables = soup.select('table')
````

````python
rows = tables[3].select('tbody > tr')
cells = []
for cell in rows:
    if cell.get('class')[0] == 'ligne1' or cell.get('class')[0] == 'ligne2':
        date_time = cell.select('td.tabev6 > b > a')[0].text.split('\xa0\xa0\xa0')
        date = date_time[0]
        time = date_time[1]
        lat_long_num = cell.select('td.tabev1')
        lat_long_letter = cell.select('td.tabev2')
        lat = f'{lat_long_num[0].text.strip()} {lat_long_letter[0].text.strip()}'
        long = f'{lat_long_num[1].text.strip()} {lat_long_letter[1].text.strip()}'
        region = cell.select('td.tb_region')[0].text.strip()
        cells.append({'date': date, 'time': time, 'latitude': lat, 'longitude': long, 'region name': region})
cells
df = pd.DataFrame(cells)
df20 = df.head(20)
df20
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
soup1 = soup.select('body > div.central-featured > div.central-featured-lang')
lang_art_list = []
for lang in soup1:
    language = lang.select('a > strong')[0].text.replace('\xa0', '')
    articles = lang.select('a > small > bdi')[0].text.replace('\xa0', '')
    lang_art_list.append(f'{language} - {articles[:-1]}')
lang_art_list
````
